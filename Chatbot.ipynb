```python
import random

# Mock stock data
stock_data = {
    'AAPL': {'price': 150.0, 'volume': 1000000},
    'GOOGL': {'price': 2800.0, 'volume': 500000},
    'MSFT': {'price': 300.0, 'volume': 800000},
    # Add more stocks as needed
}

# Function to fetch stock price
def get_stock_price(symbol):
    return stock_data.get(symbol, {}).get('price')

# Function to execute mock trade
def execute_trade(symbol, quantity, action):
    if action == 'buy':
        print(f"Bought {quantity} shares of {symbol}")
    elif action == 'sell':
        print(f"Sold {quantity} shares of {symbol}")
    else:
        print("Invalid action")

# Main function to handle user queries
def handle_query(query):
    if query.startswith('price of'):
        _, symbol = query.split('of')
        symbol = symbol.strip().upper()
        price = get_stock_price(symbol)
        if price:
            return f"The price of {symbol} is ${price:.2f}"
        else:
            return f"Sorry, I couldn't find data for {symbol}"
    elif query.startswith('buy'):
        _, symbol, quantity = query.split()
        symbol = symbol.upper()
        quantity = int(quantity)
        price = get_stock_price(symbol)
        if price:
            execute_trade(symbol, quantity, 'buy')
            return f"Bought {quantity} shares of {symbol} at ${price:.2f} each"
        else:
            return f"Sorry, couldn't execute buy order for {symbol}"
    elif query.startswith('sell'):
        _, symbol, quantity = query.split()
        symbol = symbol.upper()
        quantity = int(quantity)
        price = get_stock_price(symbol)
        if price:
            execute_trade(symbol, quantity, 'sell')
            return f"Sold {quantity} shares of {symbol} at ${price:.2f} each"
        else:
            return f"Sorry, couldn't execute sell order for {symbol}"
    else:
        return "I'm sorry, I don't understand that command"

# Mock user interaction loop
def main():
    print("Welcome to the AI trading chatbot!")
    while True:
        query = input("You: ").strip().lower()
        if query == 'exit':
            print("Exiting chatbot...")
            break
        response = handle_query(query)
        print("Bot:", response)

if __name__ == "__main__":
    main()
```

```python
class CreditDeductionBot:
    def __init__(self):
        self.credit_score = 700
        self.available_credits = 1000
    
    def check_credit_score(self):
        return f"Your current credit score is: {self.credit_score}"
    
    def deduct_credits(self, amount):
        if amount <= self.available_credits:
            self.available_credits -= amount
            return f"{amount} credits have been deducted. Available credits: {self.available_credits}"
        else:
            return "Insufficient credits to deduct."
    
    def handle_query(self, query):
        if query.startswith('check credit score'):
            return self.check_credit_score()
        elif query.startswith('deduct'):
            _, amount_str = query.split()
            amount = int(amount_str)
            return self.deduct_credits(amount)
        else:
            return "I'm sorry, I don't understand that command."

def main():
    bot = CreditDeductionBot()
    print("Welcome to the Credit Deduction Bot!")
    while True:
        query = input("You: ").strip().lower()
        if query == 'exit':
            print("Exiting bot...")
            break
        response = bot.handle_query(query)
        print("Bot:", response)

if __name__ == "__main__":
    main()
```
